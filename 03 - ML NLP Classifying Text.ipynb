### sklearn.datasets load_files
- Load text files with categories as subfolder names.

- Individual samples are assumed to be files stored a two levels folder structure such as the following:

`DATA_DIR/
    category_1_folder/
        file_1.txt file_2.txt … file_42.txt
    category_2_folder/
        file_43.txt file_44.txt …`
        
- And returns Dictionary-like object, with the following attributes:

`data`
> list of str Only present when load_content=True. The raw text data to learn.

`target`
>  ndarray The target labels (integer index).

````python
from sklearn.datasets import load_files

# for reproducibility
random_state = 0 

DATA_DIR = "./bbc/"
data = load_files(DATA_DIR, encoding="utf-8", decode_error="replace", random_state=random_state)
````

````python
type(data)
````

````python
import pandas as pd

df = pd.DataFrame(list(zip(data['data'], data['target'])), columns=['text', 'label'])
print(f"we have {len(df)} articles in our dataset")
df.head(10)
````

````python
articles = df['text'].values
len(articles)
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer
````

````python
vectorizer = TfidfVectorizer(max_df=0.5,min_df=2,stop_words='english')
````

`max_df`:
- When building the vocabulary ignore terms that have a document
    frequency strictly higher than the given threshold (corpus-specific
    stop words).
    
`min_df`: 
- When building the vocabulary ignore terms that have a document
    frequency strictly lower than the given threshold (corpus-specific
    stop words).

````python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 100, n_init = 1, verbose = True)
````

````python
km.fit(X)
````

````python
import numpy as np
np.unique(km.labels_, return_counts=True)
````

### Now we want to look at the articles in each cluster
We create a dictionary:
- Keys are cluster numbers
- Values are aggregated articles of each cluster

````python
X.shape
````

````python
len(km.labels_)
````

````python
text={}
for i,cluster in enumerate(km.labels_):
    oneDocument = articles[i]
    if cluster not in text.keys():
        text[cluster] = oneDocument
    else:
        text[cluster] += oneDocument # aggregate text in each cluster
````

### Let's find the most frequent words in each cluster

````python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from collections import defaultdict
from string import punctuation
from heapq import nlargest
import nltk 
````

````python
_stopwords = set(stopwords.words('english') + list(punctuation)+["million","billion","year","millions","billions","y/y","'s","''","``"])
 
````

````python
text[1][1:1000] # text[1] combines all the texts which belong to the cluster 1
# this is the 1st 1000 characters in cluster 1
````

````python
keywords = {} # keys are cluster numbers, vaules are the first 100 most frequent words of each cluster
counts={} # keys are cluster numbers, vaules are the frequency of most frequent words of each cluster
for cluster in range(5):
    word_sent = word_tokenize(text[cluster].lower())
    word_sent=[word for word in word_sent if word not in _stopwords]
    freq = FreqDist(word_sent) # dictionary of the frequency of each word
    keywords[cluster] = nlargest(100, freq, key=freq.get)
    counts[cluster]=freq # complete frequency distibution of all word for each cluster
````

### Let's find the top 10 keywords unique to each cluster

````python
keywords[0][:10]
````

````python
unique_keys={}
for cluster in range(5):   
    other_clusters=list(set(range(5))-set([cluster]))
    print("other_clusters: ", other_clusters)
    
    # in each loop, we have 1 maun_cluster and 4 other_clusters
    # keys_other_clusters is the list of keyword in other_clusters
    keys_other_clusters=set(keywords[other_clusters[0]]).union(set(keywords[other_clusters[1]]))\
    .union(set(keywords[other_clusters[2]])).union(set(keywords[other_clusters[3]]))
    
    # find the keywords which are only in the main_cluster
    unique=set(keywords[cluster])-keys_other_clusters
    unique_keys[cluster]=nlargest(10, unique, key=counts[cluster].get)
````
