````python
from urllib.request import urlopen # download the url page
from bs4 import BeautifulSoup # parse the url page and removes unuseful marks

def getTextWaPo(url):
    page = urlopen(url).read().decode('utf8')
    soup = BeautifulSoup(page,"lxml")
    text = ' '.join(map(lambda p: p.text, soup.find_all('article')))
    return text.encode('ascii', errors='replace').replace(b"?",b" ").decode('utf-8')
````

````python
text = getTextWaPo(articleURL)
````

````python
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from string import punctuation
````

````python
sents = sent_tokenize(text)
print(len(sents))
sents
````

````python
word_sent = word_tokenize(text.lower())
word_sent
````

````python
_stopwords = set(stopwords.words('english') + list(punctuation))
_stopwords
````

````python
word_sent=[word for word in word_sent if word not in _stopwords]
````

````python
word_sent
````

### `FreqDist` creates a dictioanry which:
- Keys are words
- Values are the frequency of those words

````python
from nltk.probability import FreqDist
freq = FreqDist(word_sent)
freq
````

### `nlargest` sorts any collection with 3 inputs:
(we need to rank the freq dictioanry keys)
- number of soterd elements
- collection to be sorted
- a function to sort elemenst

````python
from heapq import nlargest
````

````python
nlargest(10, freq, key=freq.get)
````

### `defaultdict` is a special dictonary:
- If you look up a key which is not in the dict, `defaultdict` automatically adds that key to the dictioanry

````python
from collections import defaultdict
ranking = defaultdict(int)

for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w] # sum of the importance of all words in each sentence
            
ranking # significance score of each sentence
````

### `ranking` shows the sentences based on the frequency of their words

````python
sents_idx = nlargest(4, ranking, key=ranking.get)
sents_idx # unsorted indices
````

````python
[sents[j] for j in sorted(sents_idx)] # print the sentences by the order of indices
````

````python
def summarize(text, n):
    sents = sent_tokenize(text)
    
    assert n <= len(sents) # checks whether the text has enough number of sentences
    word_sent = word_tokenize(text.lower())
    _stopwords = set(stopwords.words('english') + list(punctuation))
    
    word_sent=[word for word in word_sent if word not in _stopwords]
    freq = FreqDist(word_sent)
    
    
    ranking = defaultdict(int)
    
    for i,sent in enumerate(sents):
        for w in word_tokenize(sent.lower()):
            if w in freq:
                ranking[i] += freq[w]
             
        
    sents_idx = nlargest(n, ranking, key=ranking.get)
    return [sents[j] for j in sorted(sents_idx)]
````

````python
summarize(text,3)
````

````python
soup = BeautifulSoup(page.content, 'html.parser')
````

````python
results = soup.find(id="SearchResults")
````

````python
print(results.prettify())
````

````python
job_elems = results.find_all('section', class_='card-content')
print(type(job_elems))
````

````python
for job_elem in job_elems:
    # Each job_elem is a new BeautifulSoup object.
    # You can use the same methods on it as you did before.
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    print(title_elem)
    print(company_elem)
    print(location_elem)
    print()
````

````python
for job_elem in job_elems:
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    if None in (title_elem, company_elem, location_elem):
        continue
    
    print(title_elem.text.strip())
    print(company_elem.text.strip())
    print(location_elem.text.strip())
    print()
````

````python
text_2 = '. '.join(map(lambda p: p.text.strip(), results.find_all('h2', class_='title')))
text_2 += ' '.join(map(lambda p: p.text.strip(), results.find_all('div', class_='company')))
text_2 += ' '.join(map(lambda p: p.text.strip(), results.find_all('div', class_='location')))

text_2.encode('ascii', errors='replace').replace(b"?",b" ")
text_2
````

````python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from string import punctuation
````

````python
sents = sent_tokenize(text_2)
print(len(sents))
words = word_tokenize(text_2.lower())
print(len(words))
````

````python
english_stopwords = set(stopwords.words('english') + list(punctuation))
````

````python
words = [word for word in words if word not in english_stopwords]
print(len(words))
````

````python
from nltk.probability import FreqDist
freq_dict = FreqDist(words)
freq_dict
````

````python
from heapq import nlargest
word_rank = nlargest(7, freq_dict, key=freq_dict.get )
word_rank
````

````python
from collections import defaultdict
rankings = defaultdict(int)

for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq_dict:
            rankings[i] += freq_dict[w]
        
rankings
````

````python
sent_rank = nlargest(7, rankings, rankings.get )
sent_rank
````

````python
[sents[j] for j in sorted(sent_rank)] # print the sentences by the order of indices
````
